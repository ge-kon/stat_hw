$T$ — время, проведённое клиентом в мобильном приложении, в часах.

В методе Монте-Карло используй `n_monte_carlo=10000`.

```python
def confidence_interval(sample, confidence_level=0.95):
    z = sps.norm.ppf((1 + confidence_level) / 2)
    me = sample.sd() * z / (len(sample) ** 0.5)
    return sample.mean() - me, sample.mean() + me
```
